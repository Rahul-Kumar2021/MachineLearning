### What is missing values?



**Deletion:** Removing rows or columns with missing values. This can lead to loss of information and potentially biased results if missingness is not random.

**Imputation:** Filling in missing values with estimated or imputed values. Various methods such as mean imputation, median imputation, regression imputation, and more advanced techniques like k-nearest neighbors imputation or machine learning-based imputation can be used.

**Model-Based Approaches:** Incorporating the missing data process into the analysis model itself, allowing for more accurate parameter estimation.

**Multiple Imputation:** Creating multiple plausible imputed datasets to account for the uncertainty associated with missing data.

**Domain Knowledge:** Sometimes, domain knowledge can be used to make informed decisions about how to handle missing values.








## How to handle Missing values?
- Deleting rows with missing values(only 5% data missing then apply)
- Impute missing values for continuous variable(**ex-** 1,2,3,4,1.1,2.5{Numerical})
- Impute missing values for continuous variable(**ex-** day-night,red blue,{Categorical(**Fix number of choices**)})
- Using Alogrithms that supports Missing values 
- Prediction of missing values
- Imputation using Deep Learning Libary - Datawig
