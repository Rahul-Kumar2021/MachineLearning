```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv('E:\\DUCAT_PYTHON COURSE\\dataset\\analysis\\restaurant.csv')
df
```

```python
sb.scatterplot(x=df.total_bill,y=df.tip,hue=df.gender)
plt.xlabel('Total_Bill')
plt.ylabel('Tip')
plt.show()
```

```python
sb.scatterplot(x=df.total_bill,y=df.tip,hue=df.day)
plt.xlabel('Total_Bill')
plt.ylabel('Tip')
plt.show()
```

```python
sb.scatterplot(x=df.total_bill,y=df.tip,style=df.gender)
plt.xlabel('Total_Bill')
plt.ylabel('Tip')
plt.show()
```

```python
sb.scatterplot(x=df.total_bill,y=df.tip,hue=df.gender,style=df.day)
plt.xlabel('Total_Bill')
plt.ylabel('Tip')
plt.show()
```

```python
# Bar Chart
sb.countplot(x=df.day)
plt.show()
```

```python
sb.countplot(x=df.day,hue=df.time)
plt.show()
```

```python
sb.countplot(x=df.day,order=['Thur','Fri','Sat','Sun'])
plt.show()
```

```python
sb.countplot(y=df.day,order=['Thur','Fri','Sat','Sun'])
plt.show()
```

```python
# Joinplot===> Scatter+histogram
sb.jointplot(x=df.total_bill,y=df.tip)
plt.show()
```

```python
#pairplot===>histogram+scatter for each pair of numeric cols
sb.pairplot(df)
plt.show()
```

```python
#boxplot-->outliers detection in numeric col(s)
import numpy as np
x=np.random.randint(1,50,100)
#y=np.random.randint(1,50,100)

x[10]=150
x[20]=155
x[30]=-50
#sb.boxplot(x)
sb.boxplot(x,orient='h')
#sb.boxplot([x,y])
plt.show()
```
