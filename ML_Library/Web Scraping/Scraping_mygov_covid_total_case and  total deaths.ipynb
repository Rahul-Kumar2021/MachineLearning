# Web Scraping

```python
import requests
import bs4
```

```python
con=requests.get("https://www.mygov.in/corona-data/covid19-statewise-status/")
```

```python
data = con.text
```

```python
print(data)
```

```python
soup=bs4.BeautifulSoup(data)
```

```python
tag= soup.find('title')
print(tag.getText())
```

```python
tag = soup.find("div") # return first div tag
print(tag)
```

```python
tag= soup.find("div",{'class':'field-item even'})
print(tag)
```

```python
tags= soup.findAll("div",{'class':'field-item even'})
#print(tag)
for tag in tags:
    print(tag)
```

```python
tag_date = tags[0]
print(tag_date.getText())

tag_active = tags[2]
print("Active Cases:",tag_active.getText())

tag_deaths=tags[5]
print("Total Deaths",tag_deaths.getText())
```

```python
tags=soup.findAll("div",{'class':'field field-name-field-select-state field-type-list-text field-label-above'})
print(tags)
```

```python
state_names=[]
for tag in tags:
    state_names.append(tag.find("div",{'class':'field-item even'}).getText())
```

```python
print(state_names)
```

```python
total_cases = []
tags= soup.findAll("div",{'class':'field field-name-field-total-confirmed-indians field-type-number-integer field-label-above'})
for tag in tags:
    total_cases.append(tag.find("div",{'class':'field-item even'}).getText())
    
```

```python
print(total_cases)
```

```python
total_deaths = []
tags= soup.findAll("div",{'class':'field field-name-field-deaths field-type-number-integer field-label-above'})
for tag in tags:
    total_deaths.append(tag.find("div",{'class':'field-item even'}).getText())
```

```python
print(total_deaths)
```

```python
import pandas as pd 
```

```python
df = pd.DataFrame()
df
```

```python
df['state'] =state_names
df['total_cases']=total_cases
df['total_deaths']=total_deaths
```

```python
df
```

```python
df.index=df.state
```

```python
df
```

```python
df.drop('state',axis=1,inplace=True)
```

```python
df
```

```python
df.sort_values(by="total_cases",ascending=False)
```

```python
tags=soup.findAll("img")
for tag in tags:
    src=tag.get("src")
    url='https://www.mygov.in'+src
    con=requests.get(url)
    data=con.content
    index=url.rindex('/')
    img_name=url[index+1:]
    file= open(f"E:/myimgs/{img_name}","wb")
    file.write(data)
    file.close()
print("Img Download")
```
