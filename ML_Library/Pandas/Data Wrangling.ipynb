# Fill missing values

```python
df.fillna(20) # returns a new df after filling 20
```

```python
df.fillna({'temperature':33,'windspeed':10,'event':'unknown'})
df.replace({'temperature':'[^0-9]'},'',regex=True)
df
```

```python
df.ffill() #forward filling
```

```python
df['temperature']=df.temperature.ffill()
df['windspeed']=df.windspeed.bfill()
```
