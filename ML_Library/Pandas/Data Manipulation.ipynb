```python
def cal_tax(bill):
    if(bill<10):
        return bill*.05
    elif(bill>=10 and bill<20):
        return bill*.10
    else:
        return bill*.15
```

```python
df['tax']=df.total_bill.apply(cal_tax)
df
```
